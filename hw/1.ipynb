#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
data = pd.DataFrame()
for i in range(10):
  info = pd.read_csv(f'./Data/Students_info_{i}.csv')
  marks = pd.read_csv(f'./Data/Students_marks_{i}.csv')
  data = pd.concat([data, info.merge(marks)], axis=0)
data.head(120)
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
from numpy import mean

def DataFrameFilter(do_not_print=False):
  race_result = dict()
  education_result = dict()
  for index, row in data.iterrows():
    if row['race/ethnicity'] not in race_result:
      race_result[row['race/ethnicity']] = [row['reading score']]
    else:
      race_result[row['race/ethnicity']].append(row['reading score'])
    if row['education'] not in education_result:
      education_result[row['education']] = [row['reading score']]
    else:
      education_result[row['education']].append(row['reading score'])
  if do_not_print:
    return
  print('race/ethnicity')
  for key, values in sorted(race_result.items()):
    print(f'{key}\t\t\t{mean(values)}')
  print('education')
  for key, values in sorted(education_result.items()):
    print(f'{key}\t\t\t{min(values)}')


DataFrameFilter()
```

```python
%timeit DataFrameFilter(do_not_print=True)
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####      90 и больше = A
####      80-89 = B
####      70-79 = C
####      60-69 = D
####      50-59 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
def GetGrade(average_mark):
    if average_mark >= 90:
      return 'A'
    elif average_mark >= 80:
      return 'B'
    elif average_mark >= 70:
      return 'C'
    elif average_mark >= 60:
      return 'D'
    elif average_mark >= 50:
      return 'E'
    return 'F'

data['Grade'] = data[columns].apply(lambda x: GetGrade((x['math score'] + x['reading score'] + x['writing score']) / 3), axis=1)
data.head()
```

```python
data.groupby('Grade').count().rename(columns={'gender': 'count'})['count']
```
